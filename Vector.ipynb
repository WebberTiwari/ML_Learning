# Vector Operations

```python
#Importing libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
#Plotting vector
plt.quiver(0,0,4,5)
plt.show()
```

```python
#Setting the scale and adding grid to graph
plt.quiver(0,0,4,5, scale_units='xy', angles='xy', scale=1, color='b')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
#Plotting Multiple vector in Single graph
plt.quiver(0,0,4,5, scale_units='xy', angles='xy', scale=1, color='b')
plt.quiver(0,0,-3,-7, scale_units='xy', angles='xy', scale=1, color='r')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

# Addition of Two vector

```python
vector_1=np.asarray([0,0,2,3])

vector_2=np.asarray([0,0,3,-2])

vector_3=vector_1+vector_2

print(vector_3)
```

```python
plt.quiver(0,0,2,3, scale_units='xy', angles='xy', scale=1, color='b')
plt.quiver(0,0,3,-2, scale_units='xy', angles='xy', scale=1, color='y')
plt.quiver(0,0,5,1, scale_units='xy', angles='xy', scale=1, color='r')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

# Subtraction of Two vector

```python
vector_1=np.asarray([0,0,2,3])

vector_2=np.asarray([0,0,3,-2])

vector_3=vector_1-vector_2

print(vector_3)
```

```python
plt.quiver(0,0,2,3, scale_units='xy', angles='xy', scale=1, color='b')
plt.quiver(0,0,3,-2, scale_units='xy', angles='xy', scale=1, color='y')
plt.quiver(0,0,-1,5, scale_units='xy', angles='xy', scale=1, color='r')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

# Multiply a vector By Scaler

```python
vector_1=np.asarray([0,0,2,3])

vector_2= 2 * vector_1

print(vector_2)
```

```python
plt.quiver(0,0,2,3, scale_units='xy', angles='xy', scale=1, color='b')

plt.quiver(0,0,4,6, scale_units='xy', angles='xy', scale=1, color='r')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```
