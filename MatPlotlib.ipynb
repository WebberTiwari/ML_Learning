# MatPlotLib

```python
#import matplotlib
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x=np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(x)
```

# Plotting Curve

```python
#plotting sine curve
plt.plot(x,y)
plt.show()
```

```python
#plotting cosine curve
plt.plot(x,z)
plt.show()
```

```python
plt.plot(x,x)
plt.show()
```

```python
#Plotting curve with title, x and y axis lable
plt.plot(x,y)
plt.title('Sine Wave')
plt.xlabel('angle')
plt.ylabel('value')
plt.show()
```

```python
#plot parabola symmetric about y axis

x=np.linspace(-10,10,40)
y=x**2
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'r+')
plt.show()
```

```python
plt.plot(x,y,'g.')
plt.show()
```

```python
plt.plot(x,y,'y*')
plt.show()
```

```python
#Multiple Graph in a single  box
x=np.linspace(0,10,100)
plt.plot(x,np.sin(x), 'r+')
plt.plot(x,np.cos(x), 'g.')
plt.show()
```

## Bar Plot

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
Lang=['C','C++', 'Java', 'Python','Rust']
students=[100,80,40,90,30]
ax.bar(Lang,students)
plt.xlabel('Languages')
plt.ylabel('Number of Students')
plt.show()
```

# PIE CHART

```python
fig1=plt.figure()
ax=fig1.add_axes([0,0,1,1])
Lang=['C','C++', 'Java', 'Python','Rust']
students=[100,80,40,90,30]
ax.pie(students,labels=Lang, autopct='%1.1f%%')

plt.show()
```

```python
#Scatter plot

x = np.linspace(0, 10, 30)
y = np.sin(x)
z = np.cos(x)

# Create figure and axes
fig2 = plt.figure()
ax = fig2.add_axes([0, 0, 1, 1])

# Scatter plots
ax.scatter(x, y, color='r', label='sin(x)')  # red
ax.scatter(x, z, color='g', label='cos(x)')  # green


# Add legend
ax.legend()

plt.show()
```

# 3D Scatter PLot

```python
fig3 = plt.figure()
ax = fig3.add_subplot(111, projection='3d')  # ✅ fixed line
z = 20 * np.random.random(200)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x, y, z, c=z, cmap='Blues')
plt.show()
```
