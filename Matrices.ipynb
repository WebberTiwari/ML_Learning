## **Multiplication of Two matrices**

```python
#Random Matrices

c=np.random.randint(10, size=(3,3))

d=np.random.randint(20, size=(3,3))
```

```python
ans=np.dot(c,d)
print(ans)
```

```python
#Random Matrices

m=np.random.randint(10, size=(3,3))

n=np.random.randint(20, size=(2,3))

ans2=np.dot(m,n)
print(ans2)
```

Column of first matrix is not equal to row of second matrix

**Element** **wise** **Multiplication**

```python
#Random Matrices

c=np.random.randint(10, size=(3,3))

d=np.random.randint(20, size=(3,3))

ans=np.multiply(c,d)
#ans= c*d

print(ans)
```
