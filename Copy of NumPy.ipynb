# Some Mathematical Operations on NumPy Array


```python
#incase of List, (+) will perform concatination
list1=[1,2,3,4]
list2=[5,6,7,8]
print(list1+list2)
```

```python
#But we want to do some operation like add,sub,multiply and divide on Numpy Array

a=np.random.randint(1,10,(3,3))
b=np.random.randint(1,10,(3,3))
print(a)
print(b)
```

```python
print(a+b)
print(a-b)
print(a*b)
print(a/b)
```

```python
#Another way to do this

a=np.random.randint(1,10,(3,3))
b=np.random.randint(1,10,(3,3))
print(a)
print(b)
```

```python
print(np.add(a,b))
print(np.subtract(a,b))
print(np.multiply(a,b))
print(np.divide(a,b))
```

NumPy Array Manupulation

```python
#Taking Tranpose

a=np.random.randint(1,10,(2,3))
print(a)
print(a.shape)

Trans=np.transpose(a)
print(Trans)
print(Trans.shape)
```

```python
a=np.random.randint(1,10,(2,3))
print(a)
print(a.shape)

b=a.T
print(b)
print(b.shape)
```
